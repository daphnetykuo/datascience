```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
onshore_df = pd.read_excel("/Users/daphne/Desktop/Onshore_crm.xlsx")
onshore_df
```

```python
#將都市分入北、中、南、東四區
def which_area(city):
    city = city[:3]
    city_group = {'NORTH': ["台北市", "新北市", "基隆市", "新竹市", "桃園市", "新竹縣"], 'CENTER': ["台中市", "苗栗縣", "彰化縣", "南投縣", "雲林縣"],
                  'SOUTH': ["高雄市", "台南市", "嘉義市", "嘉義縣", "屏東縣", "澎湖縣"], 'EAST': ["宜蘭縣", "花蓮縣", "台東縣"]}

    for area in city_group:
        if city in city_group[area]:
            return area

        
#觀察客戶性別與居住地      
client_gen_area = []
i = 0
for client in onshore_df:
    gen = onshore_df['gendercode'][i]
    area = which_area(onshore_df.address1_line1[i])
    #企業不計入
    if gen == 'X':
        i += 1
        continue
    client_set = [gen, area]
    client_gen_area.append(client_set)
    i += 1

client_gen_area
```

```python
#計算每個地區的男女人數
area = ['NORTH', 'CENTER', 'SOUTH', 'EAST']
m_num = [0, 0, 0, 0]
f_num = [0, 0, 0, 0]

for clients in client_gen_area:
    i = area.index(clients[1])
    if clients[0] == 'M':
        m_num[i] += 1
    elif clients[0] == 'F':
        f_num[i] += 1

# 畫長條圖
plt.bar(area, f_num, label='Male', align='edge', width=0.3)
plt.bar(area, m_num, label='Female', align='edge', width=-0.3)
plt.legend()   #顯示label

plt.title('Onshore Clients Distribution')
plt.xlabel("Area")
plt.ylabel("Total Numbers of People")

plt.show()
```
