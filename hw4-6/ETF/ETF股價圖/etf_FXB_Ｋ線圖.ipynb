```python
import pandas as pd
import numpy as np
import datetime
from matplotlib.pylab import date2num
import matplotlib.pyplot as plt
import mpl_finance as mpf

stock = pd.read_csv('/Users/daphne/Desktop/etf_FXB.csv')
print(stock.head())
```

股價Ｋ線圖

```python
# 將資料轉換成candlestick_ohlc()方法可讀取的格式
data_list = []
for dates, row in stock.iterrows():
    # 將數字轉為時間
    date_time = datetime.datetime.strptime(row['date'], '%Y-%m-%d')
    t = date2num(date_time)
    open, high, low, close = row[1:5]
    datas = (t, open, high, low, close)
    data_list.append(datas)

# 創建子圖
fig, ax = plt.subplots()
fig.subplots_adjust(bottom=0.2)

# x軸：時間
ax.xaxis_date()
plt.xticks(rotation=45)
plt.yticks()
plt.title("Fluctuations of FXB price since 2016")
plt.xlabel('time')
plt.ylabel('share price')
mpf.candlestick_ohlc(ax, data_list, width=1.5, colorup='r', colordown='green')
plt.grid()
plt.show()
```
