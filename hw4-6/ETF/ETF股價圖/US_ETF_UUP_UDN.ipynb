## 利用 Yahoo Finance 爬取最新股價資訊

```python
#爬取 2017-now 即時股價
def crawl_price(stock_id):
    now = int(datetime.datetime.now().timestamp())+86400
    url = "https://query1.finance.yahoo.com/v7/finance/download/" + stock_id + "?period1=1483200000&period2=" + str(now) + "&interval=1d&events=history&crumb=hP2rOschxO0"
    
    response = requests.post(url)
    #io：建立虛擬檔案存取股價資訊
    f = io.StringIO(response.text)
    df = pd.read_csv(f, index_col='Date', parse_dates=['Date'])
    
    return df

#開始畫圖
plt.subplots_adjust(top=1.5, bottom=0.08, left=0.08, right=1.7, hspace=0.8,
                    wspace=0.2)

df1 = crawl_price('UUP')
plt.subplot(221)   #2：共兩列，1：共兩行，1：目前在第一格子圖
plt.title('Price of UUP')
df1.Close.plot()
plt.subplot(222)
plt.title('Volume of UUP')
df1.Volume.plot()

df2 = crawl_price('UDN')
plt.subplot(223)
plt.title('Price of UDN')
df2.Close.plot(color='darkorange')
plt.subplot(224)
plt.title('Volume of UDN')
df2.Volume.plot(color='darkorange')
```
