```python
import pickle
import re
import time
import io
import os
import json
import pandas as pd
from optparse import OptionParser
from pandas import DataFrame, read_csv
path='./data/'
```

```python
with open('bbc_trump', 'rb') as file:
    news = pickle.load(file)

print(news)
```

```python
news_df = pd.DataFrame(news)
news_df = news_df[['time', 'title', 'summary', 'doc']]
news_df
```

```python
pd.to_pickle(news_df, 'news_df.pkl')

with open('news_df.pkl', 'rb') as fout:
    allData = pickle.load(fout)
allData.head()
```

```python
corpus = allData['doc'].values.tolist()
```

```python
# TF-IDF
# coding:utf-8  
from sklearn.feature_extraction.text import CountVectorizer  
vectorizer = CountVectorizer()  
X = vectorizer.fit_transform(corpus)  
word = vectorizer.get_feature_names()  

from sklearn.feature_extraction.text import TfidfTransformer  
transformer = TfidfTransformer()  
tfidf = transformer.fit_transform(X)
```

```python
len(corpus)
```

```python
features = tfidf.toarray()
labels = allData['title'].values
labels
```

```python
# PCA
import numpy as np
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
transData = pca.fit_transform(features, labels)
type(transData)
allData['pca1'] = DataFrame(transData[:,0])
allData['pca2'] = DataFrame(transData[:,1])
allData.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(x_vars=['pca1'], y_vars=['pca2'], data=allData, size=5)
```

```python
textFrame = pd.DataFrame(features, columns = word)
textFrame.head()
len(features)
```

```python
allF = features.flatten()
nonZero = allF[np.nonzero(allF)]
m = np.median(nonZero)
```

```python
allF[np.nonzero(allF)]
```

```python
plt.hist(nonZero, bins=500)
plt.show()
m
```

```python
m = 0.1
dataset = list()
for i in range(len(features)):
    temp = textFrame.columns.values[[textFrame.iloc[i] >= m]]
    dataset.append(temp.tolist())
```

```python
for i in range(len(features)):
    print(dataset[i])
```

```python
# Apriori: https://rasbt.github.io/mlxtend/user_guide/frequent_patterns/apriori/
#!pip install mlxtend
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
graph = df.T
graph.columns = allData['title']
```

```python
graph
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

minsup = 0.0085
apriori(graph, min_support = minsup, use_colnames=True)
```

```python
result = apriori(graph, min_support = minsup)
result
```

```python
rules = association_rules(result, metric="confidence", min_threshold=0.7)
rules.head()
```

```python
rules["antecedant_len"] = rules["antecedents"].apply(lambda x: len(x))
rules.head()
```

```python
out = rules[ (rules['antecedant_len'] >= 2) &
       (rules['confidence'] > 0.75) &
       (rules['lift'] > 1.2) ]
out.head()
```

```python
support = rules['support']
confidence = rules['confidence']
plt.scatter(support, confidence, alpha=0.5, marker="*")
plt.xlabel('support')
plt.ylabel('confidence') 
plt.show()
```

```python
len(rules)
rules['antecedents'].head()
```

```python
import networkx as nx
G = nx.DiGraph()
for i in range(len(allData)):
    G.add_node(i)
for i in range(len(rules['antecedents'])):
    currentFrom = list(rules['antecedents'][i])
    currentTo = list(rules['consequents'][i])
    for j in range(len(currentFrom)):
        G.add_edge(currentFrom[j], currentTo[0])
```

```python
nodeList = allData.index[allData['doc']].tolist()
#nodeListB = allData.index[allData['docClass'] == 'environ'].tolist()

plt.figure(figsize=(40,20))

# positions for all nodes
pos = nx.fruchterman_reingold_layout(G)

# nodes
nx.draw_networkx_nodes(G, pos, nodelist=nodeList, node_color="b", node_size=1000)
#nx.draw_networkx_nodes(G, pos, nodelist=nodeListB, node_color="r", node_size=1000)
nx.draw_networkx_edges(G, pos, width=6)

# labels
nx.draw_networkx_labels(G,pos,font_size=20,font_family='sans-serif')

plt.axis('off')
plt.savefig("weighted_graph.png") # save as png
plt.show() # display
```
